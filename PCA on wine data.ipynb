#  Principal Component Analysis (PCA)
Principal Component Analysis (PCA) is a statistical technique for dimensionality reduction. It identifies the most significant patterns in data by creating new, uncorrelated variables called principal components. These components capture and prioritize variance in the data, enabling simplification, visualization, and improved efficiency in data analysis and machine learning.

```python
#  Examle of Principal Components Analysis using wine_data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

```python
wine_data = load_wine()
print(wine_data)
```

```python
# for training data
x = wine_data['data']
print(x)
```

```python
# for testing data
y = wine_data['target']
print(y)
```

```python
#  x,y = wine_data['data'],wine_data['target']
plot = plt.scatter(x[:,0],x[:,2],c=y)
plt.legend(handles=plot.legend_elements()[0],
labels = list(wine_data['target_names']))
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(x)
x_transform = pca.transform(x)
```

```python
print(x_transform)
print(x_transform.shape)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# normaloized
scaler = StandardScaler()
scaler.fit(x)
x_normalised = scaler.transform(x)
pca.fit(x_normalised)
x_transform = pca.transform(x_normalised)
```

```python
# Visulaize 
plot = plt.scatter(x_transform[:,0],x_transform[:,1],c=y)
plt.legend(handles=plot.legend_elements()[0],
labels = list(wine_data['target_names']))
plt.show()
```
